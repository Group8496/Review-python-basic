```python
missing_data = dataset.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
avg_pid = dataset["pid"].astype("float").mean(axis=0)
print("Average of pid:", avg_pid)
```

```python
dataset["pid"].replace(np.nan, avg_pid, inplace=True)
```

```python
dataset['age'].value_counts()
```

```python
dataset['age'].value_counts().idxmax()
```

```python
#replace the missing 'age' values by the most frequent 
dataset["age"].replace(np.nan, "61", inplace=True)
```

```python
# simply drop whole row with NaN in "days_to_cancer" column
dataset.dropna(subset=["days_to_cancer"], axis=0, inplace=True)

# reset index, because we droped two rows
dataset.reset_index(drop=True, inplace=True)
```
